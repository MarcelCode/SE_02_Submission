Now we will focus on the time and space complexity more in detail and will show some examples.

When analysing algorithms, it is mainly not relevant to count every step, as we did in the intro. Instead it is more relevant to get an approximation of the time complexity. Therefor you will often see these common Big-O functions for explaining algorithms:

<table class="table">
    <tbody><tr><th>Name</th><th>Big O</th></tr>

    <tr><td>Logarithmic</td><td>O(log(n))</td></tr>

</tbody></table>

In the following image, you can clearly see th etime complexity evolves from a constant O(c) to an exponential O(2^n).

### Constant Complexity (O(C))

```python
def constant(items):
    print(items[0] * 2)

constant([4, 5, 6, 8])
```

```python
input_array_size = [2, 4, 6, 8, 10, 12]
steps = [2, 2, 2, 2, 2, 2]

plt.plot(input_array_size, steps, 'b')
plt.xlabel('Inputs')
plt.ylabel('Steps')
plt.title('Constant Complexity')
plt.show()
```

### Linear Complexity (O(n))

```python
def linear(items):
    for item in items:
        print(item)

linear([4, 5, 6, 8])
```

```python
input_array_size = [2, 4, 6, 8, 10, 12]
steps = [2, 4, 6, 8, 10, 12]

plt.plot(input_array_size, steps, 'b')
plt.xlabel('Inputs')
plt.ylabel('Steps')
plt.title('Linear Complexity')
plt.show()
```

```python
def linear_twice(items):
    for item in items:
        print(item)
    
    for item in items:
        print(item)

linear_twice([4, 5, 6, 8])
```

```python
input_array_size = [2, 4, 6, 8, 10, 12]
steps = [4, 8, 12, 16, 20, 24]

plt.plot(input_array_size, steps, 'b')
plt.xlabel('Inputs')
plt.ylabel('Steps')
plt.title('Linear Complexity 2n')
plt.show()
```

### Quadratic Complexity (O(n^2))

```python
def quadratic(items):
    for item in items:
        for item2 in items:
            print(item, ' ' ,item2)

quadratic([4, 5, 6, 8])
```

```python
input_array_size = [2, 4, 6, 8, 10, 12, 14, 16, 18]
steps = [4, 16, 36, 64, 100, 144, 196, 256, 324]

plt.plot(input_array_size, steps, 'b')
plt.xlabel('Inputs')
plt.ylabel('Steps')
plt.title('Linear Complexity 2n')
plt.show()
```

## Finding the Complexity of a function with multiple operations

```python
def complex_function(items):
    print("hello")
    print("test")
    for item in items:
        print(item)
    
    for item in items:
        for item2 in items:
            print(item, item2)
    
    print("done")
    print("over and out")

    
complex_function([4, 5, 6, 8])
```

\begin{align}
O(2 + n + n^2 + 2) => O(n^2)
\end{align}

## Worst vs Best Case Complexity

```python
def linear_search(num, items):
    steps = 0
    for item in items:
        steps += 1
        if item == num:
            return steps
        
items = [2, 4, 6, 8, 10]
print(linear_search(2, items))
print(linear_search(10, items))
```

- best case:
\begin{align}
O(1)
\end{align}

- worst case:
\begin{align}
O(n)
\end{align}

## Space Complexity

```python
def squares(n):
    square_list = []
    for num in n:
        square_list.append(num * num)

    return square_list

nums = [2, 4, 6, 8, 10]
print(squares(nums))
```
