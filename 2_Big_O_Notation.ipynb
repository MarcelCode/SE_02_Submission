### Constant Complexity (O(C))

The time complexity of an algorithm is constant, when the required steps remain constant, independently from size of the input. The constant complexity can be discribed by O(c) where c can be any constant number.

In the next example you can see, that function only takes the first index of the items and multiply it by 2. Regardless of the size of the input, it will always print the first element multiplied by 2. Thus the big O notation can be described by O(1). 

In the second example we can see, that the first and the second index of the array will be multiplied by 2. It can be described by O(2). If you want to describe the function in a more general way, you would approximate the steps you would describe this function by O(1) as the growth rate stays constant.

```python
def constant(items, steps=0):
    steps += 1 # do sth.   
    return steps
    

constant([4, 5, 6, 8])
```

```python
def constant2(items, steps=0):
    steps += 1 # do sth.
    steps += 1 # do sth.
    return steps

constant2([4, 5, 6, 8])
```

The Plot represents the two function to visualize the effect of a constant complexity. On the X axis we plot the increasing size of the input array. On the y Axis we plot the steps needed for the input array.

```python
input_array_size = [2, 4, 6, 8, 10, 12]
steps = [constant(range(array_size)) for array_size in input_array_size]
steps2 = [constant2(range(array_size)) for array_size in input_array_size]

plt.plot(input_array_size, steps, 'b')
plt.plot(input_array_size, steps2, 'r')
plt.xlabel('Inputs')
plt.ylabel('Steps')
plt.title('Constant Complexity')
plt.show()
```

### Linear Complexity (O(n))

The time complexity of an algorithm is linear, when the required steps increase or decrease with the size of the input array. Linear complexity is denoted by O(n). 

In the following example we can see an example for a linear function. Throughout this notebook you will see that I count the steps inside the functions. The step counting only represents a line of code like a formula calculation or a print statement.

If we have an array size of 4, the for-loop will be executed 4 times and so on.

```python
def linear(items, steps=0):  
    for item in items:
        steps += 1 # do sth.
        
    return steps

linear([4, 5, 6, 8])
```

```python
input_array_size = [2, 4, 6, 8, 10, 12]
steps = [linear(range(array_size)) for array_size in input_array_size]

plt.plot(input_array_size, steps, 'b')
plt.xlabel('Inputs')
plt.ylabel('Steps')
plt.title('Linear Complexity')
plt.show()
```

```python
def linear_twice(items, steps=0):
    for item in items:
        steps += 1 # do sth
    
    for item in items:
        steps += 1 # do sth
        
    return steps

linear_twice([4, 5, 6, 8])
```

An important point is, that in case of huge numbers of the input, the constants become insignificant. In the above example the complexity of the algorithm is O(2n). In case of an array with an infinite size, twice of infinity is stil equal to infity and therefor can be written as O(n). 

```python
input_array_size = [2, 4, 6, 8, 10, 12]
steps_linear = [linear(range(array_size)) for array_size in input_array_size]
steps_linear_twice = [linear_twice(range(array_size)) for array_size in input_array_size]
print("Input Array Size:")
print(input_array_size)
print("Amount of Steps for linear:")
print(steps_linear)
print("Amount of Steps for linear_twice:")
print(steps_linear_twice)

plt.plot(input_array_size, steps_linear, 'b')
plt.plot(input_array_size, steps_linear_twice, 'r')
plt.xlabel('Inputs')
plt.ylabel('Steps')
plt.title('Linear Complexity 2n')
plt.show()
```

### Quadratic Complexity (O(n^2))

The time complexity of an algorithm is qudratic, when the required steps are a quadratic function of the size of the input. Quadratic complexity is denoted by O(n^2). The overall term for Quadratic complexity O(n^2) or cubic complexity 0(n^3) can be described as a polynomial where O(n^c).

```python
def quadratic(items, steps=0, print_it=False):
    for item in items:
        for item2 in items:
            steps += 1 # do sth.
            if print_it: print("Outer loop:", item, "\tInner loop:", item2, "\tSteps:", steps)
    return steps

print("Steps:", quadratic([4, 5, 6, 8], print_it=True))
```

In the above example you can see an example of a quadratic function. There are two loops involved where the first loop iterates through the list of items and then a nested inner loop again iterates through the input list. The total number of steps performed is n * n, where n is the number of items in the input array. In the print you can see that the first number is printed n times as it represents the outer loop.

```python
input_array_size = [2, 4, 6, 8, 10, 12, 14, 16, 18]
steps = [quadratic(range(array_size)) for array_size in input_array_size]
print(input_array_size)
print(steps)

plt.plot(input_array_size, steps, 'b')
plt.xlabel('Inputs')
plt.ylabel('Steps')
plt.title('Linear Complexity 2n')
plt.show()
```

## Finding the Complexity of a function with multiple operations

In The examples above I only showed small functions to better understand the basic principles behind big O notations. Now the qquestion arises what happens if the function performs multiple tasks with different time complexity?

```python
def complex_function(items, steps=0, print_it=False):
    if print_it: print("hello")
    steps += 1
    
    if print_it: print("test")
    steps += 1
    
    for item in items:
        if print_it: print(item)
        steps += 1
    
    for item in items:
        for item2 in items:
            if print_it: print(item, item2)
            steps += 1
    
    if print_it: print("done")
    steps += 1 
    
    if print_it: print("over and out")
    steps += 1
    
    return steps

    
print("Steps: ", complex_function([4, 5, 6, 8], print_it=True))
```

The function performs the following operations:
* prints "hello"
* prints "test"
* prints all items in the list
* double loop through all items in the list
* prints "done"
* prints "over and out"

All print statements are independent from the item list and therefor are __O(1)__. Printing all items in the list is a linear time complexity of __O(n)__ and a double loop is __0(n)__. Below you can see the equation for the "complex_function". It has to be noted, that the big O notation will be often used as approximation and thus the big O notation of an algorithm is equivalent to its slowest part. In this case __O(n)__.

\begin{align}
O(2 + n + n^2 + 2) => O(n^2)
\end{align}

## Worst vs Best Case Complexity

When someone wants to know about the complexity of an algorithm, the person is asking for the worst case scenario. To better understand both cases. let' have a look at the following function: 

```python
def linear_search(num, items):
    steps = 0
    for item in items:
        steps += 1
        if item == num:
            return steps
        
items = [2, 4, 6, 8, 10]
print(linear_search(2, items))
print(linear_search(10, items))
```

The function iterates through the list of elements till it finds the value, which you are looking for. As the number you are looking for could potentially be the first number in the list the best case can be described by __O(1)__. In the worst case the algorithms needs to look through the full list.

- best case:
\begin{align}
O(1)
\end{align}

- worst case:
\begin{align}
O(n)
\end{align}

## Space Complexity

When talking about the big O notation, people often refer to the time complexity. But the big O notation can also be used for the Space Complexity. In the example below a new list is initialised and all elements in the list will be added to this new list. Therefor the space complexity is __O(n)__. Usually there is a trade-off between optimal memory use and runtime performance. The more time efficiency you have, the less space efficiency you have and vice versa.

```python
def squares(n):
    square_list = []
    for num in n:
        square_list.append(num * num)

    return square_list

nums = [2, 4, 6, 8, 10]
print(squares(nums))
```
